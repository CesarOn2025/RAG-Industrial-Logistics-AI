```python
# =========================================
# STEP 4: Upload PDFs
# =========================================

from google.colab import files
uploaded = files.upload()
```

```python
# =========================================
# STEP 5: Load PDFs
# =========================================

documents = []

for file in uploaded.keys():
    loader = PyPDFLoader(file)
    documents.extend(loader.load())

print(f"Total documents loaded: {len(documents)}")
```

```python
# =========================================
# STEP 6: Split text into chunks
# =========================================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

texts = text_splitter.split_documents(documents)

print(f"Total chunks: {len(texts)}")
```

```python
# =========================================
# STEP 7: Create embeddings
# =========================================

embeddings = OpenAIEmbeddings()
```

```python
# =========================================
# STEP 8: Create vector database
# =========================================

db = Chroma.from_documents(texts, embeddings)
```

```python
# =========================================
# STEP 9: Ask a question (search only)
# =========================================

query = "What materials are used in Blucher marine systems?"

docs = db.similarity_search(query)

for i, doc in enumerate(docs):
    print(f"\nResult {i+1}:")
    print(doc.page_content[:500])
```

```python
# =========================================
# STEP 10: Generate answer with GPT
# =========================================

query = "What materials are used in Blucher marine systems?"

docs = db.similarity_search(query)

context = "\n".join([doc.page_content for doc in docs])

llm = ChatOpenAI()

response = llm.invoke(f"""
Use the following context to answer the question.

Context:
{context}

Question:
{query}
""")

print(response.content)
```
